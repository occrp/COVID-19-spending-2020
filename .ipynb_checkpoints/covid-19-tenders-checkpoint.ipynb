In this notebook, we will clean the data for analysis and explain some of the features and caveats of the data along the way. We have collected data into **two data sets**: tenders and contracts and unit prices.

## Source data sets

### 1. TENDERS
containing information on awarded tenders, contracts and companies they were awarded to, also to be found [in our Covid-19 tender tracker](XXX) (including explanation of the columns). This data is a mix of contracts and tenders. The main practical difference is that while a tender can have multiple parts - these are called *lots* - and a lot can have multiple winners, a contract usually only has one 'winner'. Please note that the data is **flat**, and this data therefore contains a lot of **duplication**. The data is structured around companies, meaning 1 row = 1 winning company. If a tender had multiple winning companies, it spans over multiple rows. Each row has an ID. We have also manually and/or semi-automatically **categorized** the tenders - to make them cross-border comparable and to filter out tenders that are not Covid-19 related at all. There might still be unrelated tenders in cases where tenders were categorized largerly automatically (for example the Russian or the UK data), even though we have manually checked the categorization for the biggest tenders. The category is in the `product` column.
### 2. UNIT PRICES
containing information on the prices per unit of a product, also to be found in the [Unit prices sheet](https://docs.google.com/spreadsheets/d/1Rx_njfOtoPF3cXWlIQqfHjr49ng-ZKjWNE9pQrR2X1M). The unit prices data have many origins. Most of the data come from Ukraine. They are filtered out from the [Prozorro](https://prozorro.gov.ua/en/tender/search/) environment. Portugese data were taken from the official sources and manually filtered and categorized by [Publico](https://www.publico.pt), Lithuanian data by the [Lithuanian government](https://vpt.lrv.lt/lt/statistika-ir-analize/sutarciu-duomenys), Czech data were filtered by [hlidacstatu.cz](https://www.hlidacstatu.cz/), Swedish data come from a FOIA request by XXX, Austrian data from sources at [Profil](https://www.profil.at/) Most of the time there is a reference to a source and a tender ID corresponding to the tender it is connected to in the TENDERS dataset.

## A note on working with tender data

Tender data is notoriously difficult to work with. The most common caveats are:
* the data is **incomplete** in many ways. Most of the countries **don't publishes all their expenses**. In some countries (e.g. Portugal, Spain, Czechia, Russia, UK) nearly all contracts are openly published. In others (e.g. Germany, France, The Netherlands) only the largest, so called EU-level tenders are published. The largest tenders account for only about ~1/3 of the total expenditures. Even if a tender is published, some of the **information is missing**. It's not unusual to see for example the price either missing, or set to 0 or even 1 euro.

* the **company names are not standardized**. Unique identifiers for companies are a long grievance of many that wish to analyze tender data. Governments fill in the company names in many ways and so make it very difficult to calculate how much a particular company is making from tenders in general.

Part of this repository is a collection of cleaning scripts called `covidtenders`. We will be using it for cleaning the data. Pandas is used for the rest.

```python
import pandas as pd

import covidtenders
from importlib import reload
reload(covidtenders)

pd.set_option("max_colwidth", 500)
pd.set_option("max_rows", 100)
pd.set_option("max_columns", 50)
pd.options.display.float_format = '{:,.2f}'.format
pd.set_option('mode.chained_assignment', None)

import warnings
warnings.filterwarnings("ignore")
```

```python
# load the whole TENDERS data set
alldata = pd.read_csv("source_tables/alldata.csv")

# Cleaning the TENDERS for TENDERS (see `covidtenders.py` for details on the steps)
TENDERS = covidtenders.filter_just_tenders(alldata)

# Cleaning the for the COMPANIES (see `covidtenders.py` for details on the steps)
COMPANIES = covidtenders.filter_just_companies(alldata,
                                               missing_prices_filled_in_dataset = "source_tables/companies_prices_missing_filled_in.csv",
                                               companies_deduped = "source_tables/companies_deduped.csv")

# Loading and cleaning UNIT PRICES (see `covidtenders.py` for details on the steps)
UNIT_PRICES = pd.read_csv("source_tables/unit_prices.csv")
UNIT_PRICES = covidtenders.clean_unitprices(UNIT_PRICES)

#UNIT_PRICES.to_csv(exports/unit-prices-clean.csv, index=False)
#TENDERS.to_csv("exports/occrp-covid19-tenders.csv", index=False)
#COMPANIES.to_csv("exports/companies_final.csv", index=False)
```
